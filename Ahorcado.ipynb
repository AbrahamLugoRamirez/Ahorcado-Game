```
import random

class palabras:
  def easy(self):
    palabras = ["hola", "sol", "pan", "sal", "aseo", "salsa", "sapo", "solo"]
    return random.choice(palabras)

  def medium(self):
    palabras = ["palabra", "excelente", "colombia", "algoritmia", "programacion"]
    return random.choice(palabras)

  def hard(self):
    palabras = ["esternocleidomastoideo ", "electroencefalografista", "desoxirribonucleico", "paralelepipedo"]
    return random.choice(palabras)

class view:
  def mostrar(self, g):
    for l in g:
        print(l, end=" ")
    print("")

  def imagen(self, num):
    imagen = [
    '''
    +---+
    |   |
        |
        |
        |
        |
        =========''', '''
    +---+
    |   |
    O   |
    |   |
        |
        |
        =========''', '''
       +---+
    |   |
    O   |
   /|   |
        |
        |
        =========''', '''
    +---+
    |   |
    O   |
   /|\  |
        |
        |
        =========''', '''
    +---+
    |   |
    O   |
   /|\  |
    |   |
        |
        =========''', '''
    +---+
    |   |
    O   |
   /|\  |
    |   |
   / \  |
        =========''', '''
'''
  ]
    return imagen[num]


while True:
  print("---Ahorcado Game---")
  print("Seleciona el metodo de juego: ")
  print("1. Easy")
  print("2. Medium")
  print("3. Hard")
  print("0. exit")

  mode = input()

  if(mode == "1"):
    palinc = palabras().easy()
  elif(mode == "2"):
    palinc = palabras().medium()
  elif(mode == "3"): 
      palinc = palabras().hard()
  elif(mode == "0"):
    print("Good Bye")
    break

  guiones = "_" * len(palinc)
  view().mostrar(guiones)
  guiones = ["_"] * len(palinc)

  intentos = 0
  vocales = 0
  n = 0
  intletras = 0
  while True:   
    if(intentos >= 5 or intletras >= 2):
      print("_________________")
      print("|---GAME OVER---|")
      print("_________________")
      break

    print("---Ahorcado Game----")
    print("Intentos: ", 5-intentos)
    print(view().imagen(intentos))

    view().mostrar(guiones)
    letra = input()
    n = n + 1
    sw = 0

    if(len(letra) == 1):
      for i,l in enumerate(palinc):
        if l == letra:
          guiones[i] = letra
          sw = 1
      view().mostrar(guiones)    
    else:
      if(letra == palinc):
        guiones[0:] = letra[0:]
        view().mostrar(guiones)
        print("______________")
        print("|---WINNER---|")
        print("______________")
        print("Numero de intentos para ganar: ", n)
        break
      else:
        print("_________________")
        print("|---GAME OVER---|")
        print("_________________")
        break
    
    if(sw == 0):
      if(letra == "a" or letra == "e" or letra == "i" or letra == "o" or letra =="u"):
        intletras = intletras + 1
      else:
        intentos = intentos + 1  
        
    if "_" not in guiones:
      print("______________")
      print("|---WINNER---|")
      print("______________")
      print("Numero de intentos para ganar: ", n)
      break

```
